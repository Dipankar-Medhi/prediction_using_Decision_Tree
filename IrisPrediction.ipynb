### **Importing the required libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### **Importing the dataset**

```python
df = pd.read_csv("Iris.csv")
df.head(5)
```

```python
df.info()
```

```python
df1 = df.drop(["Id"], axis=1)
plt.figure(figsize=(15,10))
sns.pairplot(df1)
plt.show()
```

## **Data preprocessing**

Let's split the dataset into **X** and **Y** for model training and evaluation.

```python
X = df.drop(["Species", "Id"], axis = 1 )
y = df["Species"]
```

```python
X.head()
```

```python
y.head()
```

## **Model training and evaluation**

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X,y)

print('Decision Tree Classifer Created')
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=X.columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

Now let's try to predict the Species of a sample

```python
y_pred = dtree.predict([[4.5,	3.0,	1.7,	0.2]])
print("The Species is : {}".format(y_pred[0]))
```
